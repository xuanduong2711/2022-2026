```python
Q1= df['mass'].quantile(0.25)
Q3= df['mass'].quantile(0.75)
df[(df['mass'] >= Q1) & (df['mass'] <= Q3)]
```

```python
min_man_fruit=df.groupby('fruit_name')['color_score'].agg(['min', 'max']).reset_index()
min_man_fruit
```

```python
mean_std_fruit= df.groupby('fruit_name')['color_score'].agg(['mean', 'std']).reset_index()
mean_std_fruit
```
